### Number of Alleles per SNP

This will show the number of SNPs per locus identified in the `cstacks` catalog for stacks batch 7. 

```python
cd ../stacks_b7_wgenome
```

```python
infile = open("batch_7.catalog.snps.tsv", "r")
infile.readline()
n_alleles_list = []
for line in infile:
    linelist = line.strip().split()
    alleles = linelist[6:9]
    n_alleles = len([i for i in alleles if i != "-"])
    n_alleles_list.append(n_alleles)
infile.close()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.hist(n_alleles_list, bins=np.arange(0,5,1)+0.5, color="darkcyan")
plt.suptitle("Number of Alleles per SNP, Pacific cod")
plt.xlabel("Number of Alleles")
plt.ylabel("Number of SNPs")
plt.show()
```

```python
plt.close()
```

### Number of SNPs per RADtag


```python
infile = open("batch_7.catalog.snps.tsv", "r")
infile.readline()
snps_per_locus = {}
for line in infile:
    linelist = line.strip().split()
    if linelist[2] not in snps_per_locus.keys():
        snps_per_locus[linelist[2]] = 1
    elif linelist[2] in snps_per_locus.keys():
        n_snps = snps_per_locus[linelist[2]]
        n_snps += 1
        snps_per_locus[linelist[2]] = n_snps
infile.close()
snps_counts = []
for locus in snps_per_locus.keys():
    snps_counts.append(snps_per_locus[locus])
```

```python
plt.hist(snps_counts, bins=np.arange(0,30,1)+0.5, color="darkcyan")
plt.suptitle("Number of SNPs per RADtag, Pacific cod")
plt.xlabel("Number of SNPs")
plt.ylabel("Number of RADtags")
plt.show()
```

```python
plt.close()
```

### Minor Allele Frequencies

Allele frequencies were calculated using genepop; the data was filtered for loci missing 60% + data, but nothing else. populations model `write_random_snp`. All SNPs are biallelic; MAF was calculated as an average across all populations. 

```python
infile = open("batch_7_parseINF.txt", "r")
infile.readline()

maf_list = []
import numpy as np

for line in infile:
    allele_freqs=line.strip().split()[1:]
    allele1 = allele_freqs[0::2]
    allele2 = allele_freqs[1::2]
    a1_freqs = [float(i) for i in allele1 if i != "-"]
    a2_freqs = [float(i) for i in allele2 if i != "-"]
    maf_list.append(min(np.mean(a1_freqs), np.mean(a2_freqs)))
infile.close()
```

```python
plt.hist(maf_list, bins=np.arange(0, 0.5, 0.01), color="darkcyan")
plt.suptitle("Allele Frequency of Minor Alleles, Pacific cod")
plt.xlabel("Minor Allele Frequency")
plt.ylabel("Number of SNPs")
plt.show()
```

```python
plt.close()
```
